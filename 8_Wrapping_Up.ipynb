**Mendekati Masalah Machine Learning**

Sebelum menjalankan algoritma, sangat penting untuk mempertimbangkan masalah secara luas. Algoritma hanyalah bagian kecil dari proses analisis data dan pengambilan keputusan.

**Manusia dalam Proses (Humans in the Loop)**

Pertimbangkan apakah proses Anda memerlukan keputusan instan (seperti mobil otonom) atau apakah manusia dapat dilibatkan untuk mengonfirmasi keputusan yang tidak pasti. Machine learning dapat mengotomatisasi kasus-kasus sederhana, sementara kasus yang rumit diarahkan ke pakar manusia.

**Dari Prototipe ke Produksi**

Membangun skrip analisis satu kali sangat berbeda dengan membangun sistem produksi. Dalam sistem produksi, aspek rekayasa perangkat lunak seperti keandalan (reliability), waktu jalan (runtime), dan kebutuhan memori menjadi sangat relevan


**Menguji Sistem Produksi (A/B Testing)**

Evaluasi offline menggunakan data tes yang sudah dikumpulkan hanyalah langkah pertama. Untuk sistem yang berhadapan langsung dengan pengguna, diperlukan pengujian online seperti A/B Testing. Dalam A/B testing, sebagian pengguna diberikan algoritma baru (A) dan sisanya tetap menggunakan sistem lama (B) untuk membandingkan metrik kesuksesan secara nyata

**Membangun Estimator Kustom Sendiri**

Dengan mewarisi BaseEstimator dan TransformerMixin, Anda secara otomatis mendapatkan metode seperti fit_transform secara gratis
